```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0.9)
question = '인프런에는 어떤 강의들이 있나요?'
llm.invoke(question)
```

```python
text_prompt = f'아래 질문에 답변해주세요 : {question}'
llm.invoke(text_prompt)  # invoke
```

```python
# Prompt Template 을 사용한다면, 아래와 같이 사용할 수 있다

from langchain_core.prompts import PromptTemplate


# PromptTemplate에는 f-string 이 아닌 일반 string으로 변수를 넣어야 함
prompt_with_template = '아래 질문에 답변해주세요 :\n\n {question}'

# 변수 삽입
prompt_template = PromptTemplate(template=prompt_with_template, input_variables=['question'])  

# LLM을 invoke하는게 아니다 -> 우리의 질문이 PromptTemplate에 들어가도록 하는 것
# StringPromptValue(text='아래 질문에 답변해주세요 :\n\n 인프런에는 어떤 강의들이 있나요?')
prompt_template.invoke({"question": question})

# 그리고 invoke 한 결과로 LLM을 invoke한다
llm.invoke(prompt_template.invoke({"question": question}))
```

```python
# chain을 활용하면 간결하게 만들 수 있다

prompt_chain = prompt_template | llm
prompt_chain.invoke({"question": question})  # invoke한번만으로도 처리할 수 있다
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.inflearn.com/courses/it-programming")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500, 
    chunk_overlap=200,
)

document_list = loader.load_and_split(text_splitter)
document_list
```

```python
# DB 적재

from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

embedding = OpenAIEmbeddings(model='text-embedding-3-large')

database = Chroma.from_documents(
    documents = document_list,
    embedding = embedding,
    collection_name='inflearn-lectures',
    persist_directory='./lectures')
```

```python
# 리트리버 선언

retriever = database.as_retriever(search_kwargs={"k": 1})
retriever.invoke(question)
retrieved_docs = retriever.invoke(question)
retrieved_docs
```

```python
# context : 웹 로더에서 가지고온 데이터를 넣고
# question : 질문 
## ***이 두가지 변수가 아래 rag_chain의 key로 들어가서 dict를 구성한다
rag_prompt_template = PromptTemplate(
    template='''
    You are a helpful assistant that can answer questions about the Inflearn website.
    You are given the following context:
    {context}
    Question: {question}
    ''',
    input_variables=["context", "question"]
)


rag_prompt_template.invoke({
    "context": retrieved_docs,
    "question": question
})  

rag_chain = rag_prompt_template | llm
rag_chain.invoke({
    "context": retrieved_docs,
    "question": question
})
```

```python
## 프롬프트를 불러와서 활용

from langchain_classic import hub

rlm_rag_prompt = hub.pull("rlm/rag-prompt")  # template을 불러오는 방법 -> 그대로 사용함

rlm_rag_chain = rlm_rag_prompt | llm
rlm_rag_chain.invoke({"context": retrieved_docs, "question": question})

# 위 rag_chain 과 거의 비슷한 결과가 나온다
```

```python
from langchain_core.runnables import RunnablePassthrough

# retriever 결과는 알아서 invoke 해서 context로 넣어주고
# RunnablePassthrough 를 통해서, 딕셔너리가 아니라 rag_chain.invoke(question) 로만 줘도 됨

rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | rlm_rag_prompt
    | llm
)

rag_chain.invoke(question)  # RunnablePassthrough -> dict를 활용하면 에러
```
